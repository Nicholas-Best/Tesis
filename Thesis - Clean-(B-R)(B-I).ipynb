Este script utiliza los siguientes paquetes de Python:

- **pandas (pd):** Proporciona estructuras de datos potentes para análisis y manipulación de datos, principalmente mediante DataFrames.
- **numpy (np):** Añade soporte para operaciones numéricas, arreglos y funciones matemáticas.
- **matplotlib.pyplot (plt):** Se utiliza para crear gráficos, diagramas y visualizaciones.
- **sklearn.linear_model:** Ofrece herramientas para aprendizaje automático, especialmente modelos de regresión y clasificación.
- **scipy.stats:** Proporciona funciones estadísticas como regresión lineal, distribuciones y pruebas estadísticas.
- **matplotlib.backends.backend_pdf.PdfPages:** Permite guardar múltiples figuras de matplotlib en un solo archivo PDF.
